```python
import seaborn as sns
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2})
import matplotlib as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

Load in data and create heatmap on genes identified in edgeR:

```python
data=pd.read_csv("data_combined_edgeR.csv",sep=",",index_col=0)
ax = sns.clustermap(np.log10(data+1),col_cluster=True,row_cluster=True,z_score=0, cmap = 'coolwarm',
                    col_colors=["blue"]*7+["purple"]*8) #Blue for term samples, Purple for preterm samples
ax.ax_heatmap.set_yticklabels(ax.ax_heatmap.get_yticklabels(), rotation=0)
ax.savefig("fig_3a.eps")
```
